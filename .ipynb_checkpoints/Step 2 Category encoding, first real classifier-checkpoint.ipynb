```python
trainf = pd.read_csv('train_features.csv')
traintar = pd.read_csv('train_labels.csv')

testf = pd.read_csv('test_features.csv')
```

```python
trainf['source_enc'] = trainf['source_class'].replace({'unknown':0, 'surface':0, 'groundwater':1})
testf['source_enc'] = testf['source_class'].replace({'unknown':0, 'surface':0, 'groundwater':1})
```

```python
encoder_oh = ce.OneHotEncoder(use_cat_names=True)
encoded = encoder_oh.fit_transform(trainf[['waterpoint_type_group', 'region']])
encoded_columns = encoded.columns
trainf[encoded_columns] = encoded[encoded_columns]

encoder_oh = ce.OneHotEncoder(use_cat_names=True)
encoded = encoder_oh.fit_transform(testf[['waterpoint_type_group', 'region']])
encoded_columns = encoded.columns
testf[encoded_columns] = encoded[encoded_columns]
```

```python
Xtr.head()
```

```python
features_all = trainf.columns

features_nonull = trainf.dropna(axis=1).columns

features_numeric = (trainf.drop(['id',
                                'longitude',
                                'latitude', 
                                'region_code']
                               , axis=1)
                    .select_dtypes(include='number')
                    .columns)


X = trainf[features_numeric]
y = traintar['status_ord']

Xtr, Xval, ytr, yval = train_test_split(X, y, 
                                        train_size=.80, 
                                        test_size=.20, 
                                        random_state=42)

Xte = testf[features_numeric]
```

```python
Xtr.head()
```

```python
from scipy.stats import randint
```

```python
pdist = {
    'n_estimators': randint(225,280),
    'max_depth': randint(4,8)
}

search = RandomizedSearchCV(
    estimator=RandomForestClassifier(n_jobs=-1, random_state=42),
    param_distributions=pdist,
    n_iter=5,
    n_jobs=4,
    cv=3,
    scoring='accuracy',
    verbose=10,
    return_train_score=True
)
search.fit(Xtr, ytr)
```

```python
print('Cross-Validation accuracy score:', search.best_score_)
```

```python
best = search.best_estimator_
ypred = best.predict(Xval)
accuracy_score(ypred, yval)
```

```python
ypred_test = pd.Series(best.predict(Xte))
```

```python


better_submit = pd.DataFrame({'id':Xte['id'], 'status_group':ypred_test.replace({2:'functional',
                                                           1:'functional needs repair', 
                                                           0:'non functional'})})
better_submit.to_csv('joebender_rforest1.csv', index=False)
```
