```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
import category_encoders as ce
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
trainf = pd.read_csv('train_features.csv')
traintar = pd.read_csv('train_labels.csv')

testf = pd.read_csv('test_features.csv')
```

```python
pd.options.display.max_rows = 100
pd.options.display.max_columns = 500
```

```python
trainf.shape, traintar.shape, testf.shape
```

```python
trainf['construction_year'] = pd.to_datetime(trainf['construction_year'])
testf['construction_year'] = pd.to_datetime(testf['construction_year'])
```

```python
traintar['status_bin'] = traintar['status_group'].replace({'functional':1,
                                                           'functional needs repair':0, 
                                                           'non functional': 0})
traintar['status_bool'] = traintar['status_bin'].astype(bool)

traintar['status_ord'] = traintar['status_group'].replace({'functional':2,
                                                           'functional needs repair':1, 
                                                           'non functional': 0})
```

```python
def wrangle(train, target, test):
    target['status_bin'] = target['status_group'].replace({'functional':1,
                                                           'functional needs repair':0, 
                                                           'non functional': 0})
    
    target['status_bool'] = traintar['status_bin'].astype(bool)
    
    target['status_ord'] = target['status_group'].replace({'functional':2,
                                                           'functional needs repair':1, 
                                                           'non functional': 0})
    
    
```

```python
encoder_ord = ce.OrdinalEncoder()
encoded2 = encoder_ord.fit_transform(trainf[['quantity']])
encoded2_columns = encoded2.columns
trainf['quantity'] = encoded2[encoded2_columns]

print(encoded2[1:10])

encoded2 = encoder_ord.transform(testf[['quantity']])
encoded2_columns = encoded2.columns
testf['quantity'] = encoded2[encoded2_columns]


print(encoded2.head())
```

```python
trainf.head()
```

```python
encoder_loo = ce.LeaveOneOutEncoder(verbose=10, cols=['management','source','extraction_type_group']).fit(trainf, traintar)
```

```python
encoded3 = encoder_loo.transform(trainf)
encoded3_columns = encoded3.columns
trainf[encoded3_columns] = encoded3[encoded3_columns]


encoded3 = encoder_loo.transform(testf)
encoded3_columns = encoded3.columns
testf[encoded3_columns] = encoded3[encoded3_columns]
```

```python
trainf['quality_group'].value_counts()
```

```python
trainf['source_enc'] = trainf['source_class'].replace({'unknown':0, 'surface':0, 'groundwater':1})
testf['source_enc'] = testf['source_class'].replace({'unknown':0, 'surface':0, 'groundwater':1})
```

```python
encoder_oh = ce.OneHotEncoder(use_cat_names=True)
encoded = encoder_oh.fit_transform(trainf[['waterpoint_type_group', 'region']])
encoded_columns = encoded.columns
trainf[encoded_columns] = encoded[encoded_columns]

encoded = encoder_oh.transform(testf[['waterpoint_type_group', 'region']])
encoded_columns = encoded.columns
testf[encoded_columns] = encoded[encoded_columns]
```

```python
from sklearn.preprocessing import LabelEncoder

columns = ['construction_year', 'date_recorded']

encoder_la = LabelEncoder()

for col in columns:
    encoder_la.fit(pd.concat([trainf[col], testf[col]]))
    encoded_tr =  encoder_la.transform(trainf[col])
    encoded_te = encoder_la.transform(testf[col])
    
    testf[col] = encoded_te
    trainf[col] = encoded_tr
```

```python
trainf.isnull().sum()
```

```python
features_all = trainf.columns

features_nonull = trainf.dropna(axis=1).columns

features_numeric = trainf.select_dtypes(include='number').columns


X = trainf[['longitude', 'latitude', 'quantity', 'gps_height', 'construction_year', 'region_code', 'district_code']]
y = traintar['status_ord']

Xtr, Xval, ytr, yval = train_test_split(X, y, 
                                        train_size=.80, 
                                        test_size=.20, 
                                        random_state=42)

Xte = testf[features_numeric]
```

```python
Xtr.isnull().sum(), Xtr.shape
```

```python
Xtr.dtypes
```

```python
from scipy.stats import randint
```

```python
pdist = {
    'n_estimators': randint(50,200),
    'max_depth': randint(1,30)
}

search = RandomizedSearchCV(
    estimator=RandomForestClassifier(n_jobs=-1, random_state=42),
    param_distributions=pdist,
    n_iter=100,
    n_jobs=4,
    cv=3,
    scoring='accuracy',
    verbose=10,
    return_train_score=True
)
search.fit(Xtr, ytr)
```

```python
print('Cross-Validation accuracy score:', search.best_score_)
```

```python
best = search.best_estimator_
ypred = best.predict(Xval)
accuracy_score(ypred, yval)
```

```python
ypred_test = pd.Series(best.predict(Xte))
```

```python


better_submit = pd.DataFrame({'id':Xte['id'], 'status_group':ypred_test.replace({2:'functional',
                                                           1:'functional needs repair', 
                                                           0:'non functional'})})
better_submit.to_csv('joebender_rforest1.csv', index=False)
```
